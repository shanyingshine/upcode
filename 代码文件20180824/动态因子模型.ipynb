```r
library(MARSS)
```

### 调用动态因子模型对影响gdp增速的协变量指标进行预测

```r
# 导入数据，行是时间（季度频率），列是影响gdp增速的协变量指标和gdp增速
data = read.csv('C:/Users/Administrator/Desktop/codeblocks/94-1802.csv', header=FALSE)
features1 = data
```

### 利用动态因子模型预测下一期影响gdp增速的协变量指标

```r
features1
```

```r
features1 = features1[1:10,2:29] 
```

```r
rows = nrow(features1);rows
```

```r
cols1 = ncol(features1);cols1
```

```r
line = rep(NA,rows)
```

```r
features1 = cbind(features1, line)
```

```r
features1 = as.matrix(features1)
```

```r
features1
```

### 调用动态因子模型做矩阵估计

```r
modd = MARSS(features1)
```

```r
modd1 = MARSS(y=features1,inits = modd$par)
```

```r
outt = augment.marssMLE(x=modd1,type.predict = c("observations", "states"), interval = "confidence", conf.level = 0.95)
```

```r
outt
```

```r
subdata = subset(outt, t==29);subdata
```

```r
write.csv(subdata, 'C:/users/Administrator/Desktop/codeblocks/影响gdp增速协变量预测数据_2017.1.csv')
```
