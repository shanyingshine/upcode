```python
import pandas as pd
import numpy as np
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_regression
from sklearn.linear_model import LinearRegression, Lasso
```

### 导入影响gdp增速的协变量数据

```python
df_all = pd.read_csv('C:/Users/pzc/Desktop/安徽省GDP预测/data.csv', encoding='gbk', header=None)
```

```python
df_all = df_all.T
```

```python
df_all.columns = df_all.iloc[0,:]
```

```python
df_all = df_all.drop([0])
```

```python
df_all = df_all.fillna(df_all.mean())
```

```python
df_all
```

### 导入预测出的下一期数据

```python
data_fit = pd.read_csv('C:/Users/pzc/Desktop/安徽省GDP预测/影响gdp增速协变量预测数据_2018.1.csv', encoding='gbk')
```

```python
data_fit['.rownames'] = df_all.columns[:-1]
```

```python
data_fit
```

### 方案一：先做特征选择再做回归预测

```python
nn = 5
```

```python
out = SelectKBest(f_regression, k=nn).fit(df_all.iloc[:,:-1],df_all.iloc[:,-1])
```

```python
df_all.columns[np.argsort(-out.scores_)[:nn]]
```

```python
X_new = SelectKBest(f_regression, k=nn).fit_transform(df_all.iloc[:,:-1],df_all.iloc[:,-1]);X_new
```

```python
# 列名需要对照数据输入
col = np.array(['规模以上工业增加值增速', '社会消费品零售总额','全部财政收入','固定资产投资额', '固定资产投资额其中第二产业'])
```

```python
Y = np.array(df_all.iloc[:,-1])
```

```python
model = LinearRegression()
model.fit(X_new, Y)
model.score(X_new, Y)
```

```python
x_new = np.array([data_fit.loc[[True if x in col else False for x in data_fit['.rownames']],'.fitted']]);x_new
```

```python
model.predict(x_new)
```
